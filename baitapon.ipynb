Câu 1: 
- Mở file CSV và chọn sinh viên có điểm Python >= 7
Yêu cầu: Sử dụng for và if để duyệt qua từng sinh viên trong file CSV, và in ra các sinh viên có điểm Python từ 7 trở lên.
- Thống kê điểm Python và C
Yêu cầu: Sử dụng Pandas để thống kê các giá trị lớn nhất, nhỏ nhất, trung bình và độ lệch chuẩn của điểm Python và điểm C.

```python
import csv
import pandas as pd
def filter_students(filename):
    try:
        with open(filename, mode='r', encoding='utf-8') as file:
            reader = csv.DictReader(file)
            print("Sinh viên có điểm Python >= 7:")
            for row in reader:
                if int(row['Điểm Python']) >= 7:
                    print(row['Họ Và Tên'], row['Điểm Python'])
    except FileNotFoundError:
        print("File không tồn tại")

filter_students('sinhvien.csv')

def thong_ke_diem(filename):
    try:
        df = pd.read_csv(filename)
        print("Thống kê điểm Python và C:")
        print(df[['Điểm Python', 'Điểm C']].describe())
    except FileNotFoundError:
        print("File không tồn tại")

thong_ke_diem('sinhvien.csv')
```

Câu 2: 
- Lọc sinh viên giỏi cả Python và C
Yêu cầu: Lọc ra các sinh viên có cả điểm Python và điểm C >= 7 và lưu vào file CSV mới.
- Tính trung bình điểm Python và điểm C sử dụng Numpy
Yêu cầu: Dùng Numpy để tính toán trung bình điểm của cả hai môn học.

```python
import pandas as pd

def loc_sinhvien_gioi(filename):
    try:
        df = pd.read_csv(filename)
        sinhvien_gioi = df[(df['Điểm Python'] >= 7) & (df['Điểm C'] >= 7)]
        sinhvien_gioi.to_csv('sinhvien_gioi.csv', index=False, encoding='utf-8-sig')
        print("Đã lưu danh sách sinh viên giỏi vào sinhvien_gioi.csv")
    except FileNotFoundError:
        print("File không tồn tại")

loc_sinhvien_gioi('sinhvien.csv')
import numpy as np

def tinh_trung_binh(filename):
    try:
        df = pd.read_csv(filename)
        diem_python_tb = np.mean(df['Điểm Python'])
        diem_c_tb = np.mean(df['Điểm C'])
        print(f"Điểm trung bình Python: {diem_python_tb}")
        print(f"Điểm trung bình C: {diem_c_tb}")
    except FileNotFoundError:
        print("File không tồn tại")

tinh_trung_binh('sinhvien.csv')
```

Câu 3: 
- Tìm sinh viên có điểm Python cao nhất
Yêu cầu: Sử dụng vòng for để tìm sinh viên có điểm Python cao nhất.
- Vẽ biểu đồ phân phối điểm Python và C
Yêu cầu: Sử dụng Matplotlib để vẽ biểu đồ phân phối điểm của cả hai môn.

```python
import csv

def sinh_vien_diem_cao_nhat(filename):
    try:
        max_diem = -1
        best_student = ''
        with open(filename, mode='r', encoding='utf-8') as file:
            reader = csv.DictReader(file)
            for row in reader:
                if int(row['Điểm Python']) > max_diem:
                    max_diem = int(row['Điểm Python'])
                    best_student = row['Họ Và Tên']
        print(f"Sinh viên có điểm Python cao nhất: {best_student} ({max_diem} điểm)")
    except FileNotFoundError:
        print("File không tồn tại")

sinh_vien_diem_cao_nhat('sinhvien.csv')
import pandas as pd
import matplotlib.pyplot as plt

def ve_bieu_do(filename):
    try:
        df = pd.read_csv(filename)
        plt.hist(df['Điểm Python'], bins=10, alpha=0.5, label='Điểm Python')
        plt.hist(df['Điểm C'], bins=10, alpha=0.5, label='Điểm C')
        plt.legend(loc='upper right')
        plt.title('Phân phối điểm Python và C')
        plt.xlabel('Điểm')
        plt.ylabel('Số lượng sinh viên')
        plt.show()
    except FileNotFoundError:
        print("File không tồn tại")

ve_bieu_do('sinhvien.csv')
```

Câu 4: 
- Sắp xếp sinh viên theo điểm Python
Yêu cầu: Sử dụng Pandas để sắp xếp danh sách sinh viên theo điểm Python giảm dần.
- Tính độ lệch chuẩn của điểm Python và C
Yêu cầu: Sử dụng Numpy để tính độ lệch chuẩn cho cả hai môn.

```python
import pandas as pd

def sap_xep_diem_python(filename):
    try:
        df = pd.read_csv(filename)
        df_sorted = df.sort_values(by='Điểm Python', ascending=False)
        print(df_sorted[['Họ Và Tên', 'Điểm Python']])
    except FileNotFoundError:
        print("File không tồn tại")

sap_xep_diem_python('sinhvien.csv')
import pandas as pd
import numpy as np

def tinh_do_lech_chuan(filename):
    try:
        df = pd.read_csv(filename)
        std_python = np.std(df['Điểm Python'])
        std_c = np.std(df['Điểm C'])
        print(f"Độ lệch chuẩn điểm Python: {std_python}")
        print(f"Độ lệch chuẩn điểm C: {std_c}")
    except FileNotFoundError:
        print("File không tồn tại")

tinh_do_lech_chuan('sinhvien.csv')
```

Câu 5: 
- Tìm sinh viên có cả điểm Python và C thấp nhất
Yêu cầu: Sử dụng for và if để tìm sinh viên có tổng điểm Python và C thấp nhất.
- Vẽ biểu đồ điểm sinh viên


```python
import csv

def sinh_vien_diem_thap_nhat(filename):
    try:
        min_tong_diem = float('inf')
        worst_student = ''
        with open(filename, mode='r', encoding='utf-8') as file:
            reader = csv.DictReader(file)
            for row in reader:
                tong_diem = int(row['Điểm Python']) + int(row['Điểm C'])
                if tong_diem < min_tong_diem:
                    min_tong_diem = tong_diem
                    worst_student = row['Họ Và Tên']
        print(f"Sinh viên có tổng điểm thấp nhất: {worst_student} (Tổng điểm: {min_tong_diem})")
    except FileNotFoundError:
        print("File không tồn tại")

sinh_vien_diem_thap_nhat('sinhvien.csv')
import pandas as pd
import matplotlib.pyplot as plt

def ve_bieu_do_diem_sinhvien(filename):
    try:
        df = pd.read_csv(filename)
        plt.plot(df['Họ Và Tên'], df['Điểm Python'], label='Điểm Python', marker='o')
        plt.plot(df['Họ Và Tên'], df['Điểm C'], label='Điểm C', marker='x')
        plt.xticks(rotation=90)
        plt.legend()
        plt.title('So sánh điểm Python và C')
        plt.xlabel('Sinh viên')
        plt.ylabel('Điểm')
        plt.tight_layout()
        plt.show()
    except FileNotFoundError:
        print("File không tồn tại")

ve_bieu_do_diem_sinhvien('sinhvien.csv')
```
